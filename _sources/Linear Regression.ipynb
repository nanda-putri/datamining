# Linear Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import LinearRegression
```

Datasets from make_regression

```python
from sklearn.datasets import make_regression
x_test, y_test = make_regression (n_samples = 200, n_features = 1, noise = 100)
plt.scatter(x_test,y_test)
plt.show()
```

```python
reg = LinearRegression().fit(x_test,y_test)
```

```python
reg.score(x_test,y_test)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
y_predict = reg.predict(x_test)
```

```python
plt.scatter(x_test,y_test)
plt.plot(x_test, y_predict, c = 'r', linewidth = 3)
plt.xlabel('x')
plt.ylabel('y')
```
